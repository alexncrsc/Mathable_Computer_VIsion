Primul script: Decuparea initiala a tablei de joc folosind HSV

```python
import os
import cv2
import numpy as np

# Paths
input_dir = r"C:\Users\alexn\OneDrive\Desktop\CAVA-2024-Tema1\antrenare"
output_dir = r"C:\Users\alexn\OneDrive\Desktop\CAVA-2024-Tema1\crop_1/"
os.makedirs(output_dir, exist_ok=True)

def crop_playboard(image_path, output_path, target_size=(1024, 1024)):

 
    img = cv2.imread(image_path)
    if img is None:
        print(f"Failed to load {image_path}")
        return


    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

 
    blue_lower = np.array([100, 100, 50])
    blue_upper = np.array([140, 255, 255])
    blue_mask = cv2.inRange(hsv, blue_lower, blue_upper)

    contours, _ = cv2.findContours(blue_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        print(f"Nu s-a gasit tabla {image_path}")
        return


    largest_contour = max(contours, key=cv2.contourArea)


    x, y, w, h = cv2.boundingRect(largest_contour)
    if w < 50 or h < 50:  
        print(f"S-a gasit tabla in {image_path}, dar este prea mica.")
        return

    
    cropped = img[y:y+h, x:x+w]

   
    cropped_resized = cv2.resize(cropped, target_size)

   
    cv2.imwrite(output_path, cropped_resized)
    print(f"tabla salvata s-a salvat {output_path}")


for filename in os.listdir(input_dir):
    if filename.lower().endswith((".jpg", ".png")):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)
        crop_playboard(input_path, output_path)
```

Al treilea script: Generarea Mastilor Binare care sunt o urmare a 
diferentlor intre imagini.

```python
import os
import cv2
import numpy as np


input_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/crop_2"  
output_masks_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/masks"
os.makedirs(output_masks_dir, exist_ok=True)


empty_board_path = r"C:\Users\alexn\OneDrive\Desktop\CAVA-2024-Tema1\tabla_goala.jpg"


empty_board = cv2.imread(empty_board_path, cv2.IMREAD_GRAYSCALE)
if empty_board is None:
    raise ValueError(f"Failed to load  {empty_board_path}")


def preprocess_image(image_path):

    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Failed to load {image_path}")
    if img.shape != empty_board.shape:
        img = cv2.resize(img, (empty_board.shape[1], empty_board.shape[0]))
    return img


def generate_difference_mask(img1, img2, threshold=70):

    diff = cv2.absdiff(img1, img2)
    _, mask = cv2.threshold(diff, threshold, 255, cv2.THRESH_BINARY)
    return mask



game_prefixes = ["1_", "2_", "3_", "4_"] 
for game_prefix in game_prefixes:
    print(f"Pentru jocul cu nr {game_prefix}...")
    game_images = sorted([f for f in os.listdir(input_dir) if f.startswith(game_prefix) and f.endswith(".jpg")])

    previous_image = None
    for i, filename in enumerate(game_images):
        image_path = os.path.join(input_dir, filename)
        current_image = preprocess_image(image_path)

        if i == 0:
            
            mask = generate_difference_mask(empty_board, current_image)
        else:
    
            mask = generate_difference_mask(previous_image, current_image)

       
        mask_output_path = os.path.join(output_masks_dir, f"{filename.replace('.jpg', '_mask.jpg')}")
        cv2.imwrite(mask_output_path, mask)
        print(f"Masca salvata din  {filename} catre {mask_output_path}")

   
        previous_image = current_image

print("Masti generate")
```

Al patrulea script: Detectare pozitiei si cropuirea  patratelor din grid 
cel mai mult modificate.

```python
import os
import cv2
import numpy as np

input_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/masks"
output_tiles_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/tiles/"
output_positions_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/fisiere_txt"
os.makedirs(output_tiles_dir, exist_ok=True)
os.makedirs(output_positions_dir, exist_ok=True)

grid_size = 14  # 14x14 grid

def crop_most_affected_tile(mask):


    mask_height, mask_width = mask.shape

    tile_height = mask_height // grid_size
    tile_width = mask_width // grid_size

    max_white_pixels = 0
    max_tile_coords = None


    for row in range(grid_size):
        for col in range(grid_size):
         
            y_start = row * tile_height
            y_end = (row + 1) * tile_height
            x_start = col * tile_width
            x_end = (col + 1) * tile_width
            tile = mask[y_start:y_end, x_start:x_end]

            white_pixel_count = np.sum(tile == 255)

            if white_pixel_count > max_white_pixels:
                max_white_pixels = white_pixel_count
                max_tile_coords = (row, col)

    if max_tile_coords:
        row, col = max_tile_coords
        y_start = row * tile_height
        y_end = (row + 1) * tile_height
        x_start = col * tile_width
        x_end = (col + 1) * tile_width
        cropped_tile = mask[y_start:y_end, x_start:x_end]
        return cropped_tile, (row + 1, chr(65 + col))  

    return None, None


for filename in os.listdir(input_dir):
    if filename.endswith("_mask.jpg"):
        mask_path = os.path.join(input_dir, filename)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        if mask is None:
            print(f"Failed to load {filename}")
            continue

  
        cropped_tile, position = crop_most_affected_tile(mask)

        if cropped_tile is not None and position is not None:
       
            tile_output_path = os.path.join(output_tiles_dir, f"{filename.replace('_mask.jpg', '_tile.jpg')}")
            cv2.imwrite(tile_output_path, cropped_tile)
            print(f"Salvare patrat  {filename} catre {tile_output_path}")

            position_file_path = os.path.join(output_positions_dir, f"{filename.replace('_mask.jpg', '.txt')}")
            with open(position_file_path, "w") as position_file:
                row, column = position
                position_file.write(f"{row}{column} \n")
                print(f"Salvarea pozitiei din {filename} in {position_file_path}: {row}{column}")

print("Toate imaginile au fost procesate")
```

Al 5-lea script: OCR pe noile patrate si modificare txt-urilor cu 
pozitiile.

```python
import os
import cv2
import numpy as np
import easyocr
import re


reader = easyocr.Reader(['en'])


input_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/tiles"
positions_dir = r"c:/Users/alexn/OneDrive/Desktop/CAVA-2024-Tema1/fisiere_txt"

def is_image_predominantly_white(image):

    white_pixels = np.sum(image > 200)  
    total_pixels = image.size
    return (white_pixels / total_pixels) > 0.75  

def preprocess_for_ocr(image_path):

    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if is_image_predominantly_white(img):
        img = cv2.bitwise_not(img)  

   
    img = cv2.GaussianBlur(img, (3, 3), 0) 
    _, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    resized = cv2.resize(thresh, (100, 100), interpolation=cv2.INTER_CUBIC)

    return resized

def extract_numbers(text):
 
    match = re.search(r'\b\d{1,2}\b', text)  
    return match.group() if match else "0"


for filename in os.listdir(input_dir):
    if filename.endswith(".jpg"):
        image_path = os.path.join(input_dir, filename)

     
        preprocessed_image = preprocess_for_ocr(image_path)

        
        result = reader.readtext(preprocessed_image, detail=0)

        
        detected_text = extract_numbers(result[0]) if result else "0"

   
        base_name = filename.replace("_tile.jpg", ".txt")
        position_file_path = os.path.join(positions_dir, base_name)

        if os.path.exists(position_file_path):
           
            with open(position_file_path, "r") as file:
                lines = file.readlines()

            if lines:
                lines[0] = lines[0].strip() + f" {detected_text}\n"  

           
            with open(position_file_path, "w") as file:
                file.writelines(lines)

            print(f"Adaugare la  {position_file_path}: {detected_text}")
        else:
            print(f"Failed to load {filename}")

print("Toate imaginile au fost procesate.")
 
```
